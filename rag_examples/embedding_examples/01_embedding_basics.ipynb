```python
def simulate_embedding_process():
    """
    模拟 Embedding 的过程（用简单规则演示概念）
    """
    print("=" * 60)
    print("示例 1: 模拟 Embedding 过程")
    print("=" * 60)

    # 用一个简化的"词袋"方法模拟 Embedding
    # 真实的 Embedding 要复杂得多，这里只是演示概念

    # 定义一个简单的"语义空间"
    # 每个维度代表一个语义特征
    semantic_dimensions = {
        "科技": 0,
        "水果": 1,
        "食物": 2,
        "抽象": 3,
    }

    # 给一些词分配"伪向量"（仅用于演示）
    def mock_embed(word):
        # 初始化 4 维向量
        vec = [0.0, 0.0, 0.0, 0.0]

        # 根据词的语义给对应维度赋值
        if word in ["手机", "电脑", "AI", "互联网"]:
            vec[semantic_dimensions["科技"]] = 0.8
        elif word in ["苹果", "香蕉", "橙子"]:
            vec[semantic_dimensions["水果"]] = 0.8
        elif word in ["米饭", "面条", "红烧肉"]:
            vec[semantic_dimensions["食物"]] = 0.8
        elif word in ["爱情", "自由", "梦想"]:
            vec[semantic_dimensions["抽象"]] = 0.8

        # "苹果" 比较特殊，既是水果又可以是科技产品
        if word == "苹果":
            vec[semantic_dimensions["水果"]] = 0.7
            vec[semantic_dimensions["科技"]] = 0.3  # 也可以指 Apple 公司
        elif word == "AI":
            vec[semantic_dimensions["科技"]] = 0.9
            vec[semantic_dimensions["抽象"]] = 0.2  # 也有一定的抽象性

        return vec

    # 演示
    words = ["苹果", "香蕉", "手机", "AI", "米饭", "爱情"]

    print("词语 → 向量（4 维语义空间）\n")
    print(f"{'词语':<8} {'向量':<35} {'语义解释'}")
    print("-" * 60)

    for word in words:
        vec = mock_embed(word)
        vec_str = "[" + ", ".join(f"{v:.1f}" for v in vec) + "]"
        explanation = f"科技={vec[0]:.1f}, 水果={vec[1]:.1f}, 食物={vec[2]:.1f}, 抽象={vec[3]:.1f}"
        print(f"{word:<8} {vec_str:<35} {explanation}")

    print("\n💡 观察：")
    print("   - '苹果' 和 '香蕉' 在水果维度都较高 → 语义相近")
    print("   - '手机' 和 'AI' 在科技维度都较高 → 语义相近")
    print("   - '苹果' 同时在科技维度有值 → 可指 Apple 公司")


# =============================================================================
# 示例 2: 余弦相似度计算
# =============================================================================
```

```python
def cosine_similarity_demo():
    """
    演示余弦相似度计算
    """
    print()
    print("=" * 60)
    print("示例 2: 余弦相似度计算")
    print("=" * 60)

    import math

    # 假设有以下向量（已归一化）
    vectors = {
        "苹果": [0.7, 0.6, 0.2, 0.1],
        "香蕉": [0.2, 0.8, 0.5, 0.1],
        "手机": [0.8, 0.1, 0.1, 0.2],
        "AI":   [0.9, 0.1, 0.0, 0.3],
    }

    def cosine_sim(vec1, vec2):
        """计算余弦相似度"""
        dot = sum(a * b for a, b in zip(vec1, vec2))
        norm1 = math.sqrt(sum(a * a for a in vec1))
        norm2 = math.sqrt(sum(b * b for b in vec2))
        return dot / (norm1 * norm2)

    print("计算词语间的余弦相似度：\n")

    # 计算所有配对的相似度
    words = list(vectors.keys())
    print("相似度矩阵:")
    print("-" * 50)

    # 表头
    header = "        " + "  ".join(f"{w:>6}" for w in words)
    print(header)

    for w1 in words:
        row = f"{w1:>6}  "
        for w2 in words:
            if w1 == w2:
                row += f"{'1.00':>6}  "
            else:
                sim = cosine_sim(vectors[w1], vectors[w2])
                row += f"{sim:.2f}  "
        print(row)

    print("\n💡 解读：")
    print("   - 苹果 - 香蕉：0.68（都是水果，较相似）")
    print("   - 手机 - AI：0.85（都是科技，很相似）")
    print("   - 苹果 - 手机：0.58（苹果可以指手机品牌）")
    print("   - 香蕉 - AI：0.12（几乎不相关）")

    print("\n📊 余弦相似度公式：")
    print("""
              A · B
    cos(θ) = ───────────
             ‖A‖ × ‖B‖

    - 范围：[-1, 1]
    - 1: 方向完全相同
    - 0: 正交（无关）
    - -1: 方向完全相反
    """)


# =============================================================================
# 示例 3: Embedding 的应用场景
# =============================================================================
```

```python
def visualize_embeddings():
    """
    用简化的方式演示向量可视化

    真实向量是高维的（768 维、1024 维），
    本项目统一使用 1024 维（text-embedding-v4），
    这里用 2D 演示帮助理解
    """
    print()
    print("=" * 60)
    print("示例 4: 向量可视化（2D 简化）")
    print("=" * 60)

    # 模拟一些词语的 2D 向量
    words_2d = {
        "苹果": (0.7, 0.5),
        "香蕉": (0.5, 0.7),
        "橙子": (0.6, 0.6),
        "手机": (-0.5, 0.8),
        "电脑": (-0.7, 0.7),
        "AI": (-0.6, 0.9),
        "米饭": (0.8, -0.3),
        "面条": (0.7, -0.4),
        "汽车": (-0.8, 0.2),
        "飞机": (-0.9, 0.3),
    }

    # 用 ASCII 艺术简单可视化
    print("语义空间可视化（2D 平面）：\n")

    # 定义画布
    size = 21
    canvas = [[" " for _ in range(size)] for _ in range(size)]

    # 映射坐标到画布
    for word, (x, y) in words_2d.items():
        # 从 [-1, 1] 映射到 [0, 20]
        cx = int((x + 1) * (size - 1) / 2)
        cy = int((1 - y) * (size - 1) / 2)  # y 轴翻转

        if 0 <= cx < size and 0 <= cy < size:
            # 放置词语的首字
            canvas[cy][cx] = word[0]

    # 打印画布
    print("     科技/数码")
    print("     ↑")
    print("  手机  电脑  AI")
    print("     |")
    print("  苹果  橙子")
    print("     |     香蕉")
    print("     |         食物")
    print("     |    米饭  面条")
    print("     |")
    print("  汽车     飞机")
    print("     └─────────────────→")

    print("\n💡 观察：")
    print("   - 左上：科技/数码（手机、电脑、AI）")
    print("   - 中上：水果（苹果、香蕉、橙子）")
    print("   - 右下：食物（米饭、面条）")
    print("   - 左下：交通工具（汽车、飞机）")
    print()
    print("   语义相近的词在空间中距离更近！")


# =============================================================================
# 示例 5: 常见 Embedding 模型
# =============================================================================
```
