````python
def benchmark_recommendations():
    """
    如何在自己数据上测试模型
    """
    print()
    print("=" * 60)
    print("示例 6: 实测建议")
    print("=" * 60)

    print("""
┌─────────────────────────────────────────────────────────┐
│ 如何在自己的数据上测试 Embedding 模型                       │
├─────────────────────────────────────────────────────────┤
│                                                         │
│ 步骤 1: 准备测试集                                       │
│ ────────────────────────────                            │
│ - 收集 50-100 条典型查询                                 │
│ - 为每条查询标注相关文档（人工）                        │
│ - 确保覆盖各种查询类型                                  │
│                                                         │
│ 步骤 2: 生成向量                                        │
│ ────────────────────────────                            │
│ - 用候选模型编码所有文档                                │
│ - 用候选模型编码所有查询                                │
│ - 存储向量到文件或数据库                                │
│                                                         │
│ 步骤 3: 检索测试                                        │
│ ────────────────────────────                            │
│ - 对每个查询，计算与所有文档的相似度                    │
│ - 按相似度排序，返回 Top-K                              │
│ - 计算 Recall@K 指标                                    │
│                                                         │
│ 步骤 4: 对比结果                                        │
│ ────────────────────────────                            │
│ - Recall@5: 前 5 条结果中有多少是相关的                   │
│ - Recall@10: 前 10 条结果中有多少是相关的                 │
│ - 选择 Recall 最高的模型                                │
│                                                         │
└─────────────────────────────────────────────────────────┘

💡 测试代码框架:

```python
from sentence_transformers import SentenceTransformer, util

# 1. 加载模型
model = SentenceTransformer('bge-large-zh-v1.5')

# 2. 编码文档
doc_embeddings = model.encode(documents)

# 3. 编码查询
query_embeddings = model.encode(queries)

# 4. 计算相似度
similarities = util.cos_sim(query_embeddings, doc_embeddings)

# 5. 计算 Recall@K
def recall_at_k(similarities, relevant_docs, k):
    correct = 0
    total = len(relevant_docs)
    for i, sim in enumerate(similarities):
        top_k = sim.topk(k).indices
        if relevant_docs[i] in top_k:
            correct += 1
    return correct / total

recall = recall_at_k(similarities, relevant_docs, k=5)
print(f"Recall@5: {recall:.2%}")
```
""")


# =============================================================================
# 主程序入口
# =============================================================================
````
